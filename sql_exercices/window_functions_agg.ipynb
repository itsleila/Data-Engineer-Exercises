```python
import pandas as pd
import sqlite3
```

```python
df = pd.read_csv('./data/vendas.csv')
df.head()
```

```python
connection = sqlite3.connect("./db/vendas.db")
crsr = connection.cursor()
print("Connected to the database")
```

```python
df.to_sql('vendas', connection, if_exists='replace', index=False)
print("DataFrame inserted into the database")
```

```python
'''Mostre todas as vendas com uma coluna extra:
total de vendas por vendedor
(SUM com OVER)'''
sql_query = """ SELECT *, SUM(valor) OVER(PARTITION BY vendedor) AS total_vendas_por_vendedor
FROM vendas;"""

df_total_vendas = pd.read_sql_query(sql_query, connection)
df_total_vendas.head(10)
```

```python
'''Mostre o valor de cada venda e o valor médio por região'''

sql_query = """SELECT id_venda, vendedor, regiao, valor, AVG(valor) OVER( PARTITION BY regiao) AS media_vendas_por_regiao
FROM vendas;"""

df_media_vendas_por_regiao = pd.read_sql_query(sql_query, connection)
df_media_vendas_por_regiao.head(10)
```

```python
'''quantidade de vendas feitas pelo mesmo vendedor'''
sql_query = """ SELECT *, COUNT(id_venda) OVER(PARTITION BY vendedor) AS total_vendas_por_vendedor
FROM vendas;"""

df_total_vendas_por_vendedor = pd.read_sql_query(sql_query, connection)
df_total_vendas_por_vendedor.head(10)
```

```python
'''total de vendas por região'''
sql_query = """SELECT id_venda, vendedor, regiao, valor, SUM(valor) OVER( PARTITION BY regiao) AS total_vendas_por_regiao
FROM vendas;"""

df_total_vendas_por_regiao = pd.read_sql_query(sql_query, connection)
df_total_vendas_por_regiao.head(10)
```

```python
'''valor máximo vendido por vendedor'''
sql_query = """SELECT id_venda, vendedor, regiao, valor, MAX(valor) OVER( PARTITION BY vendedor) AS max_venda_por_vendedor
FROM vendas;"""

df_max_venda_por_vendedor = pd.read_sql_query(sql_query, connection)
df_max_venda_por_vendedor.head(10)
```

```python
'''valor mínimo vendido por região'''
sql_query = """SELECT id_venda, vendedor, regiao, valor, MIN(valor) OVER( PARTITION BY regiao) AS min_venda_por_regiao
FROM vendas;"""

df_min_venda_por_regiao = pd.read_sql_query(sql_query, connection)
df_min_venda_por_regiao.head(10)
```

```python
'''diferença entre o valor de cada venda e a média de vendas por vendedor'''
sql_query = """SELECT id_venda, vendedor, regiao, valor, avg(valor) OVER( PARTITION BY vendedor) AS avg_venda_por_vendedor, valor - AVG(valor) OVER(PARTITION BY vendedor) AS diferenca_media_venda_por_vendedor
FROM vendas;"""

df_diferenca_media_por_vendedor = pd.read_sql_query(sql_query, connection)
df_diferenca_media_por_vendedor.head(10)

'''positivo → acima da média
negativo → abaixo da média'''
```

```python
'''soma acumulada de vendas por vendedor ao longo do tempo'''
sql_query = """SELECT id_venda, data_venda, vendedor, regiao, valor, SUM(valor) OVER( PARTITION BY vendedor ORDER BY data_venda ROWS BETWEEN UNBOUNDED PRECEDING AND CURRENT ROW) AS acumulo_vendas_por_vendedor
FROM vendas;"""

df_acumulo_vendas_por_vendedor = pd.read_sql_query(sql_query, connection)
df_acumulo_vendas_por_vendedor.head(10)
```

```python
'''média móvel de 2 vendas por vendedor'''
sql_query = """SELECT id_venda, data_venda, vendedor, regiao, valor, AVG(valor) OVER( PARTITION BY vendedor ORDER BY data_venda ROWS BETWEEN 1 PRECEDING AND CURRENT ROW) AS moving_average_por_vendedor
FROM vendas;"""

df_moving_average_por_vendedor = pd.read_sql_query(sql_query, connection)
df_moving_average_por_vendedor.head(10)
```

```python
'''percentual que cada venda representa do total'''
sql_query = """SELECT id_venda, vendedor, regiao, valor, SUM(valor) OVER( PARTITION BY vendedor) AS total_venda_por_vendedor, valor * 1.0 / SUM(valor) OVER(PARTITION BY vendedor) AS percentual_total_venda_por_vendedor
FROM vendas;"""

df_percentual_total_venda_por_vendedor = pd.read_sql_query(sql_query, connection)
df_percentual_total_venda_por_vendedor.head(10)
```

```python
connection.close()
print("Connection closed")
```
