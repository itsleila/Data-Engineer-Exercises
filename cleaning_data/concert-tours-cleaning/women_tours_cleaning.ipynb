### Cleaning Data - Top 20 highest-grossing tours by women of all time

```python
import pandas as pd
```

```python
df = pd.read_csv('./data/data.csv')
df.head()
```

```python
print(f'Dataframe shape: {df.shape}')
print(f'Number of missing values:\n{df.isnull().sum()}')
```

```python
df.dtypes
```

```python
df.duplicated().sum()
```

```python
df.columns.to_list()
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace('\xa0', ' ').str.replace(' ', '_').str.replace('(', '').str.replace(')', '')
df.head()
```

```python
dirty_columns = ['peak', 'all_time_peak','tour_title']
for col in dirty_columns:
  df[col] = df[col].str.replace(r'\[[^\]]*\]|[†‡*]', '', regex=True)

money_columns = ['actual_gross', 'adjusted_gross_in_2022_dollars', 'average_gross']
for col in money_columns:
  df[col] = df[col].str.replace(',', '').str.replace('$', '')

df['ref.'] = df['ref.'].str.replace('[', '').str.replace(']', '')
df.head()
```

```python

integer_columns = ['peak', 'all_time_peak', 'ref.']
for col in integer_columns:
  df[col] = pd.to_numeric(df[col], errors='coerce').astype('Int64')

float_columns = ['actual_gross', 'adjusted_gross_in_2022_dollars', 'average_gross']
for col in float_columns:
  df[col] = pd.to_numeric(df[col], errors='coerce').astype(float)

df.dtypes
```

```python
df['start_year'] = df['years'].str.split('–').str[0].str.strip()
df['end_year'] = df['years'].str.split('–').str[1].str.strip().fillna(df['start_year'])
df['tour_duration'] = df['end_year'].astype(int) - df['start_year'].astype(int) + 1
df.head()
```

*Valores ausentes em `actual_gross` e parte de `peak` foram preenchidos manualmente utilizando a mesma fonte original do dataset: [Wikipedia](https://en.wikipedia.org/wiki/List_of_highest-grossing_concert_tours_by_women#All-time_ranking).*

*A coluna `all_time_peak` não foi totalmente imputada, pois esse ranking foi posteriormente descontinuado pela própria página. Versões históricas da página também apresentavam diversos valores ausentes, indicando que a ausência desses dados é estrutural e não um erro de coleta.*

```python

Adele_tour = df[df['tour_title'].str.contains('Adele Live 2016')]
df.loc[Adele_tour.index, 'actual_gross'] = 16770000.0
Mrs_carter_tour = df[df['tour_title'].str.contains('The Mrs. Carter Show World Tour')]
df.loc[Mrs_carter_tour.index, 'actual_gross'] = 22970000.0
df.loc[Mrs_carter_tour.index, 'peak'] = 5
```

```python

Summer_carnival_tour = df[df['tour_title'].str.contains('Summer Carnival')]
df.loc[Summer_carnival_tour.index, 'peak'] = 2
df.loc[Summer_carnival_tour.index, 'rank'] = 8
Formation_tour = df[df['tour_title'].str.contains('The Formation World Tour')]
df.loc[Formation_tour.index, 'peak'] = 4
The_1989_tour = df[df['tour_title'].str.contains('The 1989 World Tour')]
df.loc[The_1989_tour.index, 'peak'] = 4
Monster_ball_tour = df[df['tour_title'].str.contains('The Monster Ball Tour')]
df.loc[Monster_ball_tour.index, 'peak'] = 4

```

```python
df.drop(columns=['all_time_peak', 'ref.'], inplace=True)
```
